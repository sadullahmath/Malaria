ImageDataGenerator sınıfı, Keras kütüphanesinde yer alan ve görüntü verilerini artırarak daha fazla veri üretmek için kullanılan bir veri artırma aracıdır.

Bu kod bloğu, ImageDataGenerator sınıfından bir örnek oluşturur ve bu örneğe veri artırma yöntemlerinin parametrelerini belirtir. Aşağıdaki parametreler kullanılır:

rotation_range: Rastgele döndürme açısı aralığı (0-180 derece arasında). width_shift_range: Rastgele yatay kaydırma aralığı, görüntü genişliğinin yüzdesi olarak belirtilir. height_shift_range: Rastgele dikey kaydırma aralığı, görüntü yüksekliğinin yüzdesi olarak belirtilir. shear_range: Rastgele kayma aralığı, kayma açısı derecesi olarak belirtilir. zoom_range: Rastgele yakınlaştırma aralığı. horizontal_flip: Rastgele yatay çevirme yapar. fill_mode: Görüntü genişletilirken veya küçültülürken oluşan boş piksellerin doldurulma yöntemi. 'nearest' yöntemi, piksellerin en yakın komşusunun rengi ile doldurulmasını sağlar. Bu parametreler, image_gen adlı bir veri artırma nesnesi oluşturur ve daha sonra modelin eğitimi sırasında bu nesne kullanılarak veri artırma yapılabilir. Bu, modelin daha fazla veriye erişmesine ve genelleştirme yapmasına yardımcı olur.

The ImageDataGenerator class is a data augmentation tool in the Keras library that is used to generate more data by augmenting the image data.

This code block creates an instance of the ImageDataGenerator class and specifies parameters for data augmentation methods to that instance. The following parameters are used:

rotation_range: Random rotation angle range (between 0-180 degrees). width_shift_range: The random horizontal scroll range is specified as a percentage of the image width. height_shift_range: The random vertical scroll range is specified as a percentage of the image height. shear_range: The random shear range is specified in degrees of shear angle. zoom_range: Random zoom range. horizontal_flip: Randomly flips horizontally. fill_mode: The method of filling the empty pixels that occur when the image is enlarged or reduced. The 'nearest' method ensures that pixels are filled with the color of their nearest neighbor. These parameters create a data augmentation object named image_gen, which can then be used to augment data during training the model. This helps the model access more data and generalize.

La classe ImageDataGenerator est un outil d'augmentation de données dans la bibliothèque Keras qui est utilisé pour générer plus de données en augmentant les données d'image.

Ce bloc de code crée une instance de la classe ImageDataGenerator et spécifie les paramètres des méthodes d'augmentation de données pour cette instance. Les paramètres suivants sont utilisés :

rotation_range : plage d'angle de rotation aléatoire (entre 0 et 180 degrés). width_shift_range : la plage de défilement horizontal aléatoire est spécifiée sous la forme d'un pourcentage de la largeur de l'image. height_shift_range : la plage de défilement vertical aléatoire est spécifiée sous la forme d'un pourcentage de la hauteur de l'image. shear_range : la plage de cisaillement aléatoire est spécifiée en degrés d'angle de cisaillement. zoom_range : plage de zoom aléatoire. horizontal_flip : retourne aléatoirement horizontalement. fill_mode : la méthode de remplissage des pixels vides qui se produit lorsque l'image est agrandie ou réduite. La méthode "le plus proche" garantit que les pixels sont remplis de la couleur de leur voisin le plus proche. Ces paramètres créent un objet d'augmentation de données nommé image_gen, qui peut ensuite être utilisé pour augmenter les données lors de la formation du modèle. Cela aide le modèle à accéder à plus de données et à généraliser.

```python
image_gen = ImageDataGenerator(rotation_range=15, 
                               width_shift_range=0.10, 
                               height_shift_range=0.10, 
                               rescale=1/255, 
                               shear_range=0.1, 
                               zoom_range=0.1, 
                               horizontal_flip=True, 
                               fill_mode='nearest')
```

ImageDataGenerator ile veri artırma işlemi, var olan veri kümesindeki görüntüleri manipüle ederek yeni görüntüler oluşturur. Bu nedenle, eğitim veri setindeki veri sayısı artmaz, ancak modelin eğitim sırasında kullanabileceği daha fazla örnek varmış gibi davranır.

Örneğin, eğer eğitim veri setinde 1000 adet görüntü varsa ve ImageDataGenerator'ı belirtilen parametrelerle kullanarak her görüntüden 10 yeni görüntü oluşturulursa, toplam eğitim verisi sayısı 10000 olur. Ancak bu yeni verileri diskte depolamak yerine, modelin eğitimi sırasında dinamik olarak üretilir ve bellekte tutulur. Bu nedenle, veri artırma işlemi eğitim verisi sayısını artırmaz, ancak modelin eğitim sırasında daha fazla veriye erişmesini sağlar.

Data augmentation with ImageDataGenerator creates new images by manipulating the images in the existing dataset. Therefore, the number of data in the training dataset does not increase, but acts as if the model has more examples that it can use during training.

For example, if there are 1000 images in the training dataset and 10 new images are generated from each image using the ImageDataGenerator with the specified parameters, the total number of training data will be 10000. But instead of storing this new data on disk, it is dynamically generated and kept in memory during the training of the model. Therefore, data augmentation does not increase the number of training data, but allows the model to access more data during training.

L'augmentation des données avec ImageDataGenerator crée de nouvelles images en manipulant les images dans l'ensemble de données existant. Par conséquent, le nombre de données dans l'ensemble de données d'apprentissage n'augmente pas, mais agit comme si le modèle avait plus d'exemples qu'il peut utiliser pendant l'apprentissage.

Par exemple, s'il y a 1 000 images dans le jeu de données d'apprentissage et que 10 nouvelles images sont générées à partir de chaque image à l'aide de ImageDataGenerator avec les paramètres spécifiés, le nombre total de données d'apprentissage sera de 10 000. Mais au lieu de stocker ces nouvelles données sur disque, il est générée dynamiquement et conservée en mémoire pendant l'apprentissage du modèle. Par conséquent, l'augmentation des données n'augmente pas le nombre de données d'apprentissage, mais permet au modèle d'accéder à davantage de données pendant l'apprentissage.

## Creating the Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dropout, Flatten, Dense, Conv2D, MaxPooling2D
```

```python
model = Sequential()

model.add(Conv2D(filters=64, kernel_size=(3,3),input_shape=image_shape, padding='same', activation='relu',))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(filters=128, kernel_size=(3,3), padding='same', activation='relu',))
model.add(Conv2D(filters=128, kernel_size=(3,3), padding='same', activation='relu',))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(filters=256, kernel_size=(3,3), padding='same', activation='relu',))
model.add(Conv2D(filters=256, kernel_size=(3,3), padding='same', activation='relu',))
model.add(Conv2D(filters=256, kernel_size=(3,3), padding='same', activation='relu',))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(filters=128, kernel_size=(3,3), padding='same', activation='relu',))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(filters=64, kernel_size=(3,3), padding='same', activation='relu',))
model.add(MaxPooling2D(pool_size=(2, 2)))


model.add(Flatten())


model.add(Dense(128))
model.add(Activation('relu'))

model.add(Dropout(0.5))

model.add(Dense(1))
model.add(Activation('sigmoid'))

model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

The image_gen.flow_from_directory() method creates a data stream for training data using a folder structure where data is classified in subdirectories in a directory on disk.

The parameters of this method are:

directory: The directory path where the training data is located. target_size: The target size at which images will be resized. color_mode: The mode of the image color channels, eg. "rgb" or "grayscale". batch_size: The size of the data chunk to be created in one go. class_mode: Classification method, eg. "binary" or "categorical". shuffle: Whether to shuffle data. seed: A predetermined initial value of the random number generator that will be used to shuffle the data. This method generates a "batch" of data each time it's called, so that the model is running with new data at each step during training. Also, data augmentation operations determined by the ImageDataGenerator are dynamically applied with each call, so the model accesses more data during training and can generalize better.

"binary" or "categorical". shuffle: Whether to shuffle data. seed: A predetermined initial value of the random number generator that will be used to shuffle the data. This method generates a "batch" of data each time it's called, so that the model is running with new data at each step during testing. Also, the data augmentation operations determined by the ImageDataGenerator are not applied to the test data because the test data must remain in their original state in order to compare with the correct results.

La méthode image_gen.flow_from_directory() crée un flux de données pour les données de test à l'aide d'une structure de dossiers dans laquelle les données sont classées dans des sous-répertoires d'un répertoire sur le disque.

Les paramètres de cette méthode sont :

répertoire : le chemin du répertoire où se trouvent les données de test. target_size : la taille cible à laquelle les images seront redimensionnées. color_mode : le mode des canaux de couleur de l'image, par ex. "rgb" ou "niveaux de gris". batch_size : la taille du bloc de données à créer en une seule fois. class_mode : méthode de classification, par ex. "binaire" ou "catégoriel". shuffle : s'il faut mélanger les données. graine : une valeur initiale prédéterminée du générateur de nombres aléatoires qui sera utilisée pour mélanger les données. Cette méthode génère un "lot" de données à chaque fois qu'elle est appelée, de sorte que le modèle s'exécute avec de nouvelles données à chaque étape du test. De plus, les opérations d'augmentation de données déterminées par ImageDataGenerator ne sont pas appliquées aux données de test car les données de test doivent rester dans leur état d'origine pour être comparées aux résultats corrects.
